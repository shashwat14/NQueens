```python
import quandl
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = quandl.get("BSE/BOM500570", returns="numpy")
```

```python
print (data.shape)
```

```python
t = data.shape[0]
print (t)
print (data[0])
```

```python
features = {"date":0, "open":1, "high":2, "low":3, "close":4, "WAP":5, "No. of shares":6, "No. of trades":7, "Total Turnover":8, "Deliverable Quantity":9, "%Deli. quantity to traded Quantity":10, "spread H-L":11, "spread C-O":12}
```

```python
stock = []
for val in data:
    stock.append(val[1])
stock = np.array(stock)
print (stock)
```

```python
%matplotlib inline
plt.plot(stock)
plt.show()
```

```python
import statistics
diffs = []
histogram = [0]*1000
for i in range(6302, t-1):
    difference = abs(stock[i+1] - stock[i])
    diffs.append(difference)
    histogram[int(difference)] += 1
plt.plot(histogram)
plt.show()
mean = statistics.mean(diffs)
variance = statistics.variance(diffs)
print ("Mean : ", mean)
print ("Variance : ", variance)
```

```python
from scipy.stats import norm
x_axis = range(0,1000)
plt.plot(x_axis, norm.pdf(x_axis,mean,variance**0.5))
```
